### TAM analysis

```python
import pandas as pd
from unidecode import unidecode

```

```python
path = '../bases_depuradas/ingreso_granular.csv' 
df = pd.read_csv(path, sep=',')
df.head()
```

```python
# Group by 'desc_ent' and 'desc_mun' and calculate the mean for the specified columns
grouped_df = df.groupby(['desc_ent', 'desc_mun'])[['cuidados', 'salud', 'atenc_ambu', 'hospital', 'medicinas']].mean()
print(grouped_df)
```

```python
# Load the censo_depurado.csv file into a DataFrame
path_censo = '../bases_depuradas/censo_depurado.csv'
df_censo = pd.read_csv(path_censo, sep=',')
df_censo.head()
```

```python
# Group by 'ENTIDAD' and 'MUNICIPIO' and calculate the sum for the specified columns
grouped_sum_df = df_censo.groupby(['ENTIDAD', 'MUNICIPIO']).sum()
grouped_sum_df = grouped_sum_df.drop('Clave_Unica', axis=1)
grouped_sum_df = grouped_sum_df.drop('SECCION', axis=1)
grouped_sum_df
```

```python
df_instituciones = pd.read_csv('datos/registros_municipales_instituciones.csv')
df_farmacias = pd.read_csv('datos/registros_municipales_farmacias.csv')
df_instituciones
df_farmacias
```

```python
df_farmacias['entidad'] = df_farmacias['entidad'].apply(unidecode)
df_farmacias['municipio'] = df_farmacias['municipio'].apply(unidecode)
df_farmacias['entidad'] = df_farmacias['entidad'].str.replace(' ', '').str.upper()
df_farmacias['municipio'] = df_farmacias['municipio'].str.replace(' ', '').str.upper()
df_instituciones['entidad'] = df_instituciones['entidad'].apply(unidecode)
df_instituciones['municipio'] = df_instituciones['municipio'].apply(unidecode)
df_instituciones['entidad'] = df_instituciones['entidad'].str.replace(' ', '').str.upper()
df_instituciones['municipio'] = df_instituciones['municipio'].str.replace(' ', '').str.upper()
```

```python
df_farmacias.rename(columns={'numero_registros': 'numero_de_farmacias'}, inplace=True)
df_instituciones.rename(columns={'número_registros': 'numero_de_instituciones'}, inplace=True)
```

```python
df_mapeo = pd.read_csv("datos/mapeo_clave_nombre.csv")
df_mapeo
```

```python
df_tam = df_mapeo.copy()
df_demo = grouped_sum_df.copy()
df_granular = grouped_df.copy()
```

```python
df_demo_reset = df_demo.reset_index()
df_demo_reset.head()
```

```python
df_granular = df_granular.reset_index()
df_granular.head()
```

```python
df_tam['entidad'] = df_tam['entidad'].apply(unidecode)
df_tam['municipio'] = df_tam['municipio'].apply(unidecode)
df_tam['entidad'] = df_tam['entidad'].str.replace(' ', '').str.upper()
df_tam['municipio'] = df_tam['municipio'].str.replace(' ', '').str.upper()
df_granular['desc_ent'] = df_granular['desc_ent'].str.replace(' ', '').str.upper()
df_granular['desc_mun'] = df_granular['desc_mun'].str.replace(' ', '').str.upper()
```

```python
df_tam_merged = df_tam.merge(df_demo_reset, left_on=['cve_mun', 'cve_ent'], right_on=['MUNICIPIO', 'ENTIDAD'], how='left')
df_tam = df_tam_merged.drop(['ENTIDAD', 'MUNICIPIO'], axis=1)
df_tam
```

```python
df_tam_merged = df_tam.merge(df_granular, left_on=['municipio', 'entidad'], right_on=['desc_mun', 'desc_ent'], how='left')
df_tam = df_tam_merged.drop(['desc_mun', 'desc_ent'], axis=1)
df_tam
```

```python
df_tam = df_tam.merge(df_farmacias, left_on=['municipio', 'entidad'], right_on=['municipio', 'entidad'], how='left')
df_tam
```

```python
df_tam = df_tam.merge(df_instituciones, left_on=['municipio', 'entidad'], right_on=['municipio', 'entidad'], how='left')
df_tam
```

```python
df_tam['TAM'] = df_tam['Poblacion_Total'] * df_tam['medicinas']
df_tam
```

```python
## min min_habitantes
rh = 0.06/1000
df_tam['phalc'] = df_tam['Poblacion_Total'] * rh
fph = 4.9
df_tam['Farm_alc'] = df_tam['phalc']/fph
df_tam['Pers_Farm'] = df_tam['Poblacion_Total'] / df_tam['Farm_alc']
```

```python
df_tam
```

```python
## mayor gasto en tipos de salud .75
## precencia de instituciones de salud .25

df_tam_filtered = df_tam[df_tam['Poblacion_Total'] >= 10000]

gasto_promedio = df_tam_filtered['salud'].mean()
instituciones_promedio = df_tam_filtered['numero_de_instituciones'].mean()

df_tam_filtered['SAM'] = df_tam_filtered['TAM'] * (
    (0.75 * df_tam_filtered['salud']/gasto_promedio + 0.25 * df_tam_filtered['numero_de_instituciones']/instituciones_promedio) / 100
)

df_tam_filtered['SAM']
```

```python
df_tam_filtered
```

```python
df_tam_filtered['SOM'] = df_tam_filtered['SAM'] / (df_tam_filtered['numero_de_farmacias']+1)
df_tam_filtered
```

```python
# Select the relevant columns and sort by 'SOM'
df_ordered = df_tam_filtered[['municipio', 'entidad', 'Poblacion_Total', 'numero_de_farmacias', 'numero_de_instituciones', 'SOM']].sort_values(by='SOM', ascending=False)

# Display the ordered DataFrame
df_ordered.head(15)
```
