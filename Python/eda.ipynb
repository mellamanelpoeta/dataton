```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Set the style for seaborn
sns.set_style("whitegrid")

# Create a color palette
colors = sns.color_palette("Set2")

# Replace the codes with descriptive names and get value counts
pharmacy_counts = farmacias['codigo_act'].replace({
    464111: 'Sin minisuper', 
    464112: 'Con minisuper'
}).value_counts()

# Create the plot
plt.figure(figsize=(12, 6))
ax = sns.barplot(x=pharmacy_counts.index, y=pharmacy_counts.values, palette=colors)

# Customize the plot
plt.title('Distribución de Farmacias por Tipo', fontsize=20, pad=20)
plt.xlabel('Tipo de Farmacia', fontsize=14, labelpad=10)
plt.ylabel('Número de Farmacias', fontsize=14, labelpad=10)

# Rotate x-axis labels for better readability
plt.xticks(rotation=0, fontsize=12)
plt.yticks(fontsize=12)

# Add value labels on top of each bar
for i, v in enumerate(pharmacy_counts.values):
    ax.text(i, v + 0.5, f'{v:,}', ha='center', va='bottom', fontsize=12)

# Adjust layout and display the plot
plt.tight_layout()

# Save the plot as a high-resolution image
#plt.savefig('farmacia_tipo_distribucion.png', dpi=300, bbox_inches='tight')

# Show the plot
plt.show()

# Print the data for reference
print(pharmacy_counts)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Set the style for seaborn
sns.set_style("whitegrid")

# Create a color palette
colors = sns.color_palette("Set2")

# Group by 'entidad' and 'codigo_act', then count
grouped = farmacias.groupby(['entidad', 'codigo_act']).size().unstack(fill_value=0)

# Replace codes with descriptive names
grouped.columns = grouped.columns.map({464111: 'Sin minisuper', 464112: 'Con minisuper'})

# Calculate total pharmacies per state and sort
total_pharmacies = grouped.sum(axis=1)
sorted_states = total_pharmacies.sort_values(ascending=False).index

# Reorder the DataFrame based on the sorted states
grouped_sorted = grouped.loc[sorted_states]

# Create a stacked bar plot
fig, ax = plt.subplots(figsize=(15, 10))
grouped_sorted.plot(kind='bar', stacked=True, ax=ax, color=colors)

# Customize the plot
plt.title('Distribución de Tipos de Farmacias por Estado', fontsize=20, pad=20)
plt.xlabel('Estado', fontsize=14, labelpad=10)
plt.ylabel('Número de Farmacias', fontsize=14, labelpad=10)
plt.legend(title='Tipo de Farmacia', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=12)

# Add total numbers on top of each bar
for i, total in enumerate(total_pharmacies.loc[sorted_states]):
    ax.text(i, total, f'{total:,}', ha='center', va='bottom', fontsize=10, rotation=45)

# Adjust layout
plt.tight_layout()

# Save the plot
#plt.savefig('distribucion_farmacias_por_estado.png', dpi=300, bbox_inches='tight')

# Show the plot
plt.show()

# If you want to see the data behind the plot
print(grouped_sorted)
```

```python
# Filter for pharmacies
pharmacies_with_minisuper = gdf[gdf['codigo_act'] == '464111']
pharmacies_without_minisuper = gdf[gdf['codigo_act'] == '464112']

# Plot
fig, ax = plt.subplots(figsize=(15, 15))
gdf.plot(ax=ax, color='lightgrey', markersize=1, alpha=0.1)
pharmacies_with_minisuper.plot(ax=ax, color='red', markersize=5, label='Farmacia con minisuper')
pharmacies_without_minisuper.plot(ax=ax, color='blue', markersize=5, label='Farmacia sin minisuper')

plt.title('Location of Existing Pharmacies in Mexico')
plt.legend()
plt.axis('off')
plt.show()

# Basic statistics
print(f"Total economic units: {len(gdf)}")
print(f"Total pharmacies with minisuper: {len(pharmacies_with_minisuper)}")
print(f"Total pharmacies without minisuper: {len(pharmacies_without_minisuper)}")
print(f"Total pharmacies: {len(pharmacies_with_minisuper) + len(pharmacies_without_minisuper)}")
print(f"Pharmacy density: {(len(pharmacies_with_minisuper) + len(pharmacies_without_minisuper)) / len(gdf):.4f}")

# Distribution of pharmacy types
print("\nDistribution of pharmacy types:")
print(f"Pharmacies with minisuper: {len(pharmacies_with_minisuper) / (len(pharmacies_with_minisuper) + len(pharmacies_without_minisuper)):.2%}")
print(f"Pharmacies without minisuper: {len(pharmacies_without_minisuper) / (len(pharmacies_with_minisuper) + len(pharmacies_without_minisuper)):.2%}")
```
